```
# we then load in the data for manipulation
# we start with the Telecom datasets
# we start with the first one
T = pd.read_csv('/content/Telcom_dataset.csv')

T
```

```
# we then drop the unnecessary columns 
T3 = T.drop(columns=['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B',])
T3
```

```
T3.columns = ["PRODUCT","VALUE","DATE_TIME","CELL_ON_SITE","CELL_ID","SITE_ID"]
T3
```

```
# we then load in the second one
T1 = pd.read_csv('/content/Telcom_dataset2.csv')

T1
```

```
# drop unnecessary columns
T4 = T1.drop(columns=['DW_A_NUMBER', 'DW_B_NUMBER','COUNTRY_A','COUNTRY_B'])
T4
```

```
T4.columns = ["PRODUCT","VALUE","DATE_TIME","CELL_ON_SITE","CELL_ID","SITE_ID"]
T4
```

```
# And the final Telecom dataset
T2 = pd.read_csv('/content/Telcom_dataset3.csv')

T2
```

```
# drop the unnecessary columns.

T5 = T2.drop(columns=['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B'])
T5
```

```
# we then merge the three data sets to ease manipulation
T6= [T3,T4,T5]

T7= pd.concat(T6)
T7.head(5)
```
